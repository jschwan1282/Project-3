```python
import pandas as pd
import numpy as np
```

```python
country_csv = "malaria-deaths-by-region.csv"
age_csv = "malaria-deaths-by-age.csv"
gdp_csv = "malaria-prevalence-vs-gdp-per-capita.csv"

country = pd.read_csv(country_csv)
age = pd.read_csv(age_csv)
gdp = pd.read_csv(gdp_csv)

gdp.head()
```

```python
merged = pd.merge(country, age, how='left', on=['country', 'year'])
merged.head()
```

```python
merged2 = pd.merge(merged, gdp, how='left', on=['country', 'year'])
merged2.head()
```

```python
cleaned = merged2.drop(['code_y', 'code', ], axis=1)
cleaned.head()
```

```python
cleaned.rename(columns={'code)':'code'}, inplace=True)
```

```python
cleaned.head()
```

```python
cleaned = cleaned[['country', 'code', 'year', 'deaths', 'under_5', '5_14', '15_49', '50_69', 'over_70', 'gdp_per_capita', 'population', 'perc_of_pop_w_malaria']]
cleaned.head()
```

```python
cleaned.round({'deaths': 0, 'under_5': 0, '5_14': 0, '15_49': 0, '50_69': 0, 'over_70': 0, 'gdp_per_capita': 0, 'perc_of_pop_w_malaria': 2})
cleaned.head()
```

```python
cleaned.to_csv("merged_cleaned.csv", header=True, index=False)
```
